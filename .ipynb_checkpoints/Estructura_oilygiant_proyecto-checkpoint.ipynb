## 3. Entrenamiento y Evaluación del Modelo

### 3.1 Definición de Funciones del Pipeline

```python
def entrenar_evaluar_modelo(df, nombre_region, random_state=42):
    """
    Entrena un modelo de regresión lineal y evalúa su desempeño.
    
    Retorna:
        predicciones (array), respuestas_reales (Series), rmse (float), media_predicha (float)
    """
    # Separar features y target
    features = df[['f0', 'f1', 'f2']]
    target = df['product']
    
    # División train/validación 75/25
    X_train, X_val, y_train, y_val = train_test_split(
        features, target, test_size=0.25, random_state=random_state
    )
    
    # Entrenamiento
    modelo = LinearRegression()
    modelo.fit(X_train, y_train)
    
    # Predicciones
    predicciones = modelo.predict(X_val)
    
    # Métricas
    rmse = mean_squared_error(y_val, predicciones) ** 0.5
    media_predicha = predicciones.mean()
    
    print(f'\n========== {nombre_region} ==========')
    print(f'Volumen medio predicho: {media_predicha:.2f} miles de barriles')
    print(f'RMSE: {rmse:.2f}')
    
    return predicciones, y_val, rmse, media_predicha
```

### 3.2 Región 0 — `geo_data_0.csv`

#### 3.2.1–3.2.4 Entrenamiento, Predicciones y Métricas

```python
pred_0, real_0, rmse_0, media_0 = entrenar_evaluar_modelo(df0, 'Región 0', RANDOM_STATE)
```

## 5. Cálculo de Ganancia por Región

### 5.1 Función de Cálculo de Ganancia

```python
def calcular_ganancia(predicciones, valores_reales, n_pozos=200,
                      ingreso_por_unidad=4500, presupuesto=100_000_000):
    """
    Selecciona los n_pozos con mayor predicción y calcula la ganancia
    usando los valores reales de reservas.
    
    Retorna:
        ganancia (float) en USD
    """
    pred_series = pd.Series(predicciones, index=valores_reales.index)
    top_indices = pred_series.nlargest(n_pozos).index
    volumen_total = valores_reales[top_indices].sum()
    ingreso_total = volumen_total * ingreso_por_unidad
    ganancia = ingreso_total - presupuesto
    return ganancia
```

### 5.2 Selección de los 200 Mejores Pozos y Cálculo de Ganancia por Región

```python
ganancia_0 = calcular_ganancia(pred_0, real_0)
ganancia_1 = calcular_ganancia(pred_1, real_1)
ganancia_2 = calcular_ganancia(pred_2, real_2)

for nombre, ganancia in zip(['Región 0', 'Región 1', 'Región 2'],
                             [ganancia_0, ganancia_1, ganancia_2]):
    print(f'{nombre}: Ganancia potencial = ${ganancia:,.0f} USD')
```

## 6. Análisis de Riesgos con Bootstrapping

### 6.1 Definición de la Función de Bootstrapping

```python
def bootstrapping_region(predicciones, valores_reales, n_muestras=1000,
                         n_pozos=200, ingreso_por_unidad=4500,
                         presupuesto=100_000_000, random_state=42):
    """
    Aplica bootstrapping para estimar la distribución de ganancias.
    
    Retorna:
        ganancias (array de n_muestras)
    """
    rng = np.random.default_rng(random_state)
    pred_series = pd.Series(predicciones, index=valores_reales.index)
    ganancias = []

    for _ in range(n_muestras):
        muestra_pred = pred_series.sample(n=500, replace=True, random_state=rng.integers(0, 10000))
        muestra_real = valores_reales[muestra_pred.index]
        top_indices = muestra_pred.nlargest(n_pozos).index
        volumen = muestra_real[top_indices].sum()
        ganancia = volumen * ingreso_por_unidad - presupuesto
        ganancias.append(ganancia)

    return np.array(ganancias)


def resumen_bootstrapping(ganancias, nombre_region):
    """Imprime métricas clave del análisis de bootstrapping."""
    media      = ganancias.mean()
    ic_low     = np.percentile(ganancias, 2.5)
    ic_high    = np.percentile(ganancias, 97.5)
    riesgo     = (ganancias < 0).mean()

    print(f'\n========== {nombre_region} ==========')
    print(f'Beneficio promedio:        ${media:>15,.0f} USD')
    print(f'IC 95%:                    ${ic_low:>15,.0f} — ${ic_high:,.0f} USD')
    print(f'Riesgo de pérdidas:        {riesgo * 100:.2f}%')
    print(f'¿Pasa el umbral (<2.5%)?:  {"✅ Sí" if riesgo < 0.025 else "❌ No"}')

    return {'region': nombre_region, 'media': media, 'ic_low': ic_low,
            'ic_high': ic_high, 'riesgo': riesgo, 'ganancias': ganancias}
```

### 6.2 Región 0 — Distribución de Beneficios

```python
boot_0 = bootstrapping_region(pred_0, real_0, random_state=RANDOM_STATE)
res_0  = resumen_bootstrapping(boot_0, 'Región 0')
```

### 6.3 Región 1 — Distribución de Beneficios

```python
boot_1 = bootstrapping_region(pred_1, real_1, random_state=RANDOM_STATE)
res_1  = resumen_bootstrapping(boot_1, 'Región 1')
```

### 6.4 Región 2 — Distribución de Beneficios

```python
boot_2 = bootstrapping_region(pred_2, real_2, random_state=RANDOM_STATE)
res_2  = resumen_bootstrapping(boot_2, 'Región 2')
```

### 6.5 Comparativa de Riesgo entre Regiones

```python
resultados = [res_0, res_1, res_2]

df_resultados = pd.DataFrame([
    {
        'Región': r['region'],
        'Beneficio Promedio (USD)': f"${r['media']:,.0f}",
        'IC 95% Inferior (USD)': f"${r['ic_low']:,.0f}",
        'IC 95% Superior (USD)': f"${r['ic_high']:,.0f}",
        'Riesgo de Pérdida (%)': f"{r['riesgo'] * 100:.2f}%",
        'Cumple criterio <2.5%': '✅' if r['riesgo'] < 0.025 else '❌'
    } for r in resultados
])

print(df_resultados.to_string(index=False))
```

```python
# Visualización de distribuciones de beneficio
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

for ax, boot, res in zip(axes, [boot_0, boot_1, boot_2], resultados):
    ax.hist(boot, bins=50, color='steelblue', edgecolor='white', alpha=0.8)
    ax.axvline(0, color='red', linestyle='--', linewidth=1.5, label='Pérdida = 0')
    ax.axvline(res['media'], color='green', linestyle='-', linewidth=1.5, label='Media')
    ax.set_title(res['region'])
    ax.set_xlabel('Ganancia (USD)')
    ax.set_ylabel('Frecuencia')
    ax.legend(fontsize=8)

plt.suptitle('Distribución de Beneficios — Bootstrapping (1000 muestras)', fontsize=14)
plt.tight_layout()
plt.show()
```
